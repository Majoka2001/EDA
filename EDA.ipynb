```python
zomato_ratings=merged_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
zomato_ratings
```

**Key Observations from the ratings data of Zomato:**


1.   Ratings between 1.8 to 2.4 are 'Poor' and indicate customer dissatisfaction.
2.   Ratings between 2.5 to 3.4 are 'Average'.
3.   Ratings ranging from 3.5 to 3.9 can be considered as 'Good'.
4.   Ratings from 4.0 to 4.4 are 'Very Good' indicating a high level of Customer Satisfaction.
5.   Ratings from 4.5 to 4.9 are considered 'Excellent' signifying the highest level of Customer Satisfaction.



```python
Ratings_by_color=merged_df.groupby('Rating color').size().rename('Rating Count')
```

```python
Ratings_by_color
```

```python
User_reviews=Ratings_by_color.reset_index()
```

```python
User_reviews
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=zomato_ratings, hue='Rating color',
  palette={'White':'Blue', 'Red':'Red', 'Orange':'Orange', 'Yellow':'Yellow', 'Green':'Green', 'Dark Green':'darkgreen'}, estimator=sum)
```

```python
bins = [1.0, 1.8, 2.4, 3.4, 3.9, 4.4, 4.9]
labels = ['1.0-1.8', '1.9-2.4', '2.5-3.4', '3.5-3.9', '4.0-4.4', '4.5-4.9']

zomato_ratings['Rating Bin'] = pd.cut(zomato_ratings['Aggregate rating'], bins=bins, labels=labels, include_lowest=True)

sns.histplot(data=zomato_ratings, x='Rating Bin', stat='probability', kde=False)

plt.xlabel('Rating Bins')
plt.ylabel('Count of Ratings')
plt.title('Probability of each user rating')
plt.show()
```

```python
rating_mapping = {
    'Dark Green': 'Excellent', 'Green': 'Very Good', 'Orange': 'Average', 'Red': 'Poor', 'White': 'Not Rated', 'Yellow': 'Good'}
User_reviews['User Reviews']=User_reviews['Rating color'].map(rating_mapping)
```

```python
plt.pie(User_reviews['Rating Count'], labels=User_reviews['User Reviews'], autopct='%1.1f%%', labeldistance=1.1, pctdistance=0.6, shadow=True,
        startangle=30)
```

```python
#count plot
sns.countplot(x='Rating color', data=zomato_ratings,
              palette={'White':'Blue', 'Red':'Red', 'Orange':'Orange', 'Yellow':'Yellow', 'Green':'Green', 'Dark Green':'darkgreen'},
              stat='probability')
```

```python

# Sample DataFrame
data = {
    'Product': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Region': ['North', 'South', 'East', 'North', 'South', 'East', 'North', 'South', 'East'],
    'Sales': [200, 150, 180, 220, 210, 250, 170, 160, 200]
}

df = pd.DataFrame(data)

# Create a bar plot with hue
sns.barplot(x='Product', y='Sales', data=df, hue='Region')

# Set labels and title
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales by Product and Region')

# Show plot
plt.show()
```
